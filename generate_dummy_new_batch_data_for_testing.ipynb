```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pathlib
from tqdm import tqdm
from feature_class import *
from helper_functions import *

reference_version = "20240925"
data_version = "20240924"

path_to_feature_dir = f"/Users/hieunguyen/data/WGS_features/{data_version}/feature"
path_to_metadata = f"/Users/hieunguyen/data/WGS_features/{data_version}/metadata/metadata_nonBS_200924.xlsx"
obj = WGS_GW_features(path_to_feature_dir=path_to_feature_dir,
                      path_to_metadata=path_to_metadata)

path_to_ot_ref = f"feature_drift_reference/OT/{reference_version}"
ref_metadata = pd.read_csv(os.path.join(path_to_ot_ref, "reference_samples.csv"))

metadata = obj.match_metadata.copy()
```

```python
test_runs = [item for item in metadata.columns if "RUN_metadata test" in item and "BS" not in item]
```

```python
dummy_batch_path = "/Users/hieunguyen/data/WGS_features/dummy_new_batch"
for col in test_runs:
    os.system(f"mkdir -p {dummy_batch_path}/{col.replace(' ', '_')}")
    selected_metadata = metadata[metadata[col] == 1]
    selected_metadata["Foldername"] = selected_metadata["Filename"].apply(lambda x: f"{x.split('_GWfeature')[0]}_features")
    selected_metadata["Folderpath"] = selected_metadata["Foldername"].apply(lambda x: f"{path_to_feature_dir}/{x}")
    for p in selected_metadata.Folderpath.unique():
        os.system(f"rsync -avh --progress {p} {dummy_batch_path}/{col.replace(' ', '_')}")
```
